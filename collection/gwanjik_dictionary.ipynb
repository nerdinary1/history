```python
import csv
import json
import re
```

```python
#html to csv and csv to json
t=[]
with open('gwanjiktable.csv','r') as f:
    table=csv.DictReader(f)
    for row in table:
        gwanjik={}
        for fieldname in row:
            if fieldname=='관직명':
                gwanjik['department']=row['관직명'] 
            gwanjik['grade']=row
        gwanjik['grade'].pop('관직명')
        t.append(gwanjik)
print(t)
with open('gwanjiktable.json','w') as f:
    json.dump(t,f,ensure_ascii=False)
```

```python
#change department name 
with open('gwanjiktable.json','r') as f:
    table=json.load(f)
for dept in table:
    #treating department name
    deptname=dept['department']
    rc=re.compile(r'[^ \.\,\?\!a-zA-Z0-9\u3131-\u3163\uac00-\ud7a3]+')
    hanja=rc.match(deptname).group().strip('()][').replace('(','')
    hangul=rc.sub("",deptname)
    if len(hanja) !=len(hangul):
        hangul=hangul[1:-1]
    newname=hangul+"("+hanja+")"
    dept['department']=newname
    
    
    #treating grade value
    for grade in dept['grade']:
        gradenames=dept['grade'][grade].split(' ')
        newnames=[]
        for gradename in gradenames:
            rc=re.compile(r'[^ \.\,\?\!a-zA-Z0-9\u3131-\u3163\uac00-\ud7a3]+')
            hanja="" if rc.match(gradename) == None else rc.match(gradename).group().strip('()][').replace('(','')
            hangul=rc.sub("",gradename)
            hangul = ''.join([i for i in hangul if not i.isdigit()])
            newname=hangul+"("+hanja+")"
            if newname =="()":
                newname=""
            newnames.append(newname)
        dept['grade'][grade]=newnames

        
with open('gwanjiktable.json','w') as f:
    json.dump(table,f,ensure_ascii=False)
```

```python
#change key value order
with open('gwanjiktable.json','r') as f:
    table=json.load(f)
    for dept in table:
        ndc={}
        for grade in dept['grade']:
            for i in dept['grade'][grade]:
                if i=='':
                    continue

                if "정" in grade:
                    newgrade = 20-2*int(grade[1])
                elif "종" in grade:
                    newgrade = 19-2*int(grade[1])

                ndc[i]=newgrade
        dept['grade']=ndc

with open('gwanjikdictionary_by_buseo.json','w') as f:
    json.dump(table,f,ensure_ascii=False)
print(table)
        
                
```

```python
#merge buseo and gwanjik name
with open('gwanjikdictionary_by_buseo.json','r') as f:
    table=json.load(f)
ndic={}
for dept in table:
    deptname = dept['department']
    for gradename in dept['grade']:
        grname = gradename
        level = dept['grade'][grname]
        nname = deptname+' '+grname
        ndic[nname]=level
with open('gwanjikdictionary.json','w') as f:
    json.dump(ndic,f,ensure_ascii=False)
print(ndic)
        
```
